```python
import pandas as pd
import numpy as np
```

```python
# Multi-Indexing
df = pd.DataFrame(
    {"a": [4,5,6,6,np.nan],
    "b": [7,8,9,9,np.nan],
    "c": [10,11,12,12,np.nan]},
    index = pd.MultiIndex.from_tuples(
    [('d',1),('d',2),('e',2),('e',3),('e',5)],
    names=['n','v'])
)
df
```

```python
# Drop Duplicate
df = df.drop_duplicates(keep='last')
df
```

```python
# Row Selection > < >= <=
df[df['c'] > 7]
```

```python
# Row Selection !=  ==
df[df["b"] != 7]
```

```python
df['a'].isin([5])
```

```python
pd.isnull(df)
df['a'].isnull()
```

```python
pd.notnull(df)
```
